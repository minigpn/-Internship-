```python
from bs4 import BeautifulSoup
import requests
```

```python
page =requests.get("https://www.nobroker.in/flats-for-sale-in-sampangiram_nagar_bangalore")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
House_Title= []

for i in soup.find_all('span',class_="overflow-hidden overflow-ellipsis whitespace-nowrap max-w-80pe po:max-w-full"):
        House_Title.append(i.text)
House_Title
```

```python
Location= []

for i in soup.find_all('div',class_="mt-0.5p overflow-hidden overflow-ellipsis whitespace-nowrap max-w-70 text-gray-light leading-4 po:mb-0 po:max-w-95"):
        Location.append(i.text)
Location
```

```python
add_info =[]
aditional_info = soup.find_all('div', class_='font-semi-bold heading-6')
for i in aditional_info:
    add_info.append(i.get_text().replace("\n",""))
add_info
```

```python
Area = []
for i in range(0,74,3):
    Area.append(aditional_info[i].get_text().replace("\n",""))
Area
```

```python
EMI = []
for i in range(1,74,3):
    EMI.append(aditional_info[i].get_text().replace("\n",""))
EMI
```

```python
Price = []
for i in range(2,75,3):
    Price.append(aditional_info[i].get_text().replace("\n",""))
Price
```

```python
print(len(House_Title),len(Location),len(Area),len(Price),len(EMI ))
```
