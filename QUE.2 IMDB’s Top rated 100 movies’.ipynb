```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
movieName=[]
for i in soup.find_all('h3',class_="lister-item-header"):
    movieName.append((i.text).replace("\n",""))
movieName
```

```python
Rating=[]
for i in soup.find_all('div',class_="inline-block ratings-imdb-rating"):
    Rating.append((i.text).replace("\n",""))
Rating
```

```python
ReleaseYear=[]
for i in soup.find_all('span',class_="lister-item-year text-muted unbold"):
    ReleaseYear.append(i.text)
ReleaseYear
print(*ReleaseYear, sep='\n')
```

```python
print(len(movieName),len(Rating),len(ReleaseYear))
```

```python
import pandas as pd
df =pd.DataFrame({'movieName':movieName,'Rating':Rating,'ReleaseYear':ReleaseYear})
df
```
